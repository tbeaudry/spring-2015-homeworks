## Step 3. Advanced filtering & joins

By now, you should have noticed that, for each day, there are multiple records in our dataframe. Use the `groupby` function to get, for each day, the counts for all the case categories. Plot the result for the top 5 categories as you see fit. **(20 pts)**

````python
# Write your code here
import numpy as np
boston = pd.read_csv('cases_2015.csv',parse_dates=['OPEN_DT','TARGET_DT','CLOSED_DT'])
boston.OPEN_DT=boston.OPEN_DT.apply(lambda d: datetime.date(d))
gains = {}
for gain, gain_data in boston.groupby(["OPEN_DT"]):
    gains[gain] = gain_data
#print gains
days={}
for x in gains:
    day={'Request for Snow Plowing':0,'Snow Removal':0,'Requests for Street Cleaning':0,'Request for Pothole Repair':0,'Misc. Snow Complaint':0}
    for z in gains[x]['CASE_TITLE']:#.count('Request for Snow Plowing')
        for d in day:
            if z==(d):
                day[d]+=1
    days[x]=day

daysx=[]
c1=[]
c2=[]
c3=[]
c4=[]
c5=[]
for title,data_dict in days.iteritems():
    x=title
    daysx.append(x)
    label=data_dict.keys()
    #print label
    y=data_dict.values()
    c1.append(y[0])
    c2.append(y[1])
    c3.append(y[2])
    c4.append(y[3])
    c5.append(y[4])
idx=np.argsort(daysx)
daysx=np.array(daysx)[idx]
c1=np.array(c1)[idx]
c2=np.array(c2)[idx]
c3=np.array(c3)[idx]
c4=np.array(c4)[idx]
c5=np.array(c5)[idx]
#print daysx,c1
plt.plot(daysx,c1,label='Misc. Snow Complaint')
plt.plot(daysx,c2,label='Requests for Street Cleaning')
plt.plot(daysx,c3,label= 'Request for Pothole Repair')
plt.plot(daysx,c4,label='Request for Snow Plowing')
plt.plot(daysx,c5,label='Snow Removal')
plt.legend(label)
plt.xlabel('Date')
plt.ylabel('Number of Calls')
plt.show()
````

Let's see how this trend compares to the snowfall. First, we need to read the data from `snow.csv`

````python
# Read the dataframe
snow_data = pd.read_csv("snow.csv",parse_dates=[1])
````

According to the documentation, the column `SNOW` has the values for the snowfall, in mm. However, the dataset we have has information from many different stations around Boston. Aggregate all their measurements and transform the result to centimeters. Finally, plot the result **(10 pts)**

````python
# Write your code here
datex=[]
snowfall=[]
for snow, snowdata in snow_data.groupby(["DATE"]):
    num=0
    vol=0
    datex.append(snow)
    for x in snowdata["SNOW"]:
        num+=1
        if x>0:
            vol+=x
    mean=(vol*1.0)/(num*1.0)
    snowfall.append(mean)

plt.plot(datex,snowfall,label='Snowfall in cm')

plt.legend(label)
plt.xlabel('Date')
plt.ylabel('snowfall')
plt.show()
````

The last thing we will do is to join the snowfall dataframe with that of the top 5 categories of calls, per day. Don't forget to fill the NaN values with 0.
````python
fig, axes = plt.subplots(nrows=2,figsize=(12,5), sharex=True)

calls_on_snow.ix[:,0:5].plot(ax=axes[0])
calls_on_snow.ix[:,5].plot(ax=axes[1],label="snowfall (in cm)", color='gray')

axes[0].set_title("The effect of the winter snowstorms on Boston's 311 Hotline for 2015")
axes[0].set_ylabel("# Calls")
axes[1].set_ylabel("Snowfall in cm")
axes[1].set_xlabel("Date")
axes[0].xaxis.grid(False)
axes[1].xaxis.grid(False)
````
