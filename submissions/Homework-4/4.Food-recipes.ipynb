```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
import numpy as np
import scipy as sp
import sklearn.datasets as sk_data
import sklearn.metrics as metrics
from sklearn.cluster import KMeans
import scipy.sparse.linalg as linalg
import scipy.cluster.hierarchy as hr

import sklearn.cluster as cluster

from sklearn.decomposition import TruncatedSVD

from scipy.spatial.distance import pdist, squareform
from sklearn.preprocessing import StandardScaler

#import matplotlib as mpl
import seaborn as sns
%matplotlib inline


lines = np.loadtxt("epic_recipes.txt", dtype=str, comments="#", delimiter="\n", unpack=False)
#temp1=lines[0].split('\t',1)[0]
#temp2=' '.join(lines[0].split('\t',1)[1].split("\t"))
print lines.shape
i=0
data=[]
for x in lines:
    temp1=lines[i].split('\t',1)[0]
    temp2=' '.join(lines[i].split('\t',1)[1].split("\t"))
    temp=[]
    temp.append(temp1)
    temp.append(temp2)
    data.append(temp)
    i+=1
#print data[0]
data=np.array(data)

#Here the data is parsed into a np array of two coluns per recipe
#The first column is the country of origin and the second is a string containing the recipe
```

```python
'''
Hypothesis: Region of Origin will likely cluster fairly well with ingredients because different regions will likely share 
certain uncommon ingredients that will distance themselves from other clusters
   
    
'''
```

```python
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer()
corpus = []
for x in data:
    corpus.append(x[1])
X = vectorizer.fit_transform(corpus)
X
```

```python
Xdense=X.todense()
Xsmall=Xdense[0:500]
Xsmall.shape
```

```python
#distanceMatrix = pdist(X.todense(),'cityblock')
Z = hr.linkage(Xdense, method='complete', metric='cityblock')
print Z.shape,X.shape
fig = plt.figure(figsize=(30,30))
T = hr.dendrogram(Z,color_threshold=0.4, leaf_font_size=4)
fig.show()
```

```python
origins={}
for i in data:
    if i[0] not in origins:
        origins[i[0]]=0
    else:
        origins[i[0]]+=1
print origins, len(origins)
#Here I am mostly interested in the number of regions so i can find the hierarchal clustering with the same number
    
```
